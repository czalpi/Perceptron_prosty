## Bramka logiczna AND

obraz_2021-02-25_214528.png

````python
import numpy as np

#dane wejściowe
data = np.array([
                 [1,1,1], # [bias,x1,x2]
                 [1,1,0],
                 [1,0,1],
                 [1,0,0]
])

#dane wyjściowe
t = np.array([1,0,0,0])
````

Funkcja aktywacji 

````python
def activation_f(x):
    return 1 if x >= 0 else 0
````

````python
def fit_perceptron(x_data, target): # jako dane wejściowe podajemy zbiór treningowy oraz dane wyjściowe
    w = np.zeros(len(x_data[0])) # tworzymy wektor wag o długości równej liczbie wejść oraz dodatkowo jedno miejsce na bias
                                 # i ustawiamy ich wartości początkowe na 0
    lr = 1 # współczynnik uczenia
    epochs = 100 # liczba epok

    for epoch in range(epochs): # iterujemy po wszystkich epokach
        for i, x in enumerate(x_data): # w każdej z epok iterujemy po wszystkich przykładach treningowych

            sum = np.dot(x_data[i], w) # liczymy iloczyn skalarny dwóch wektorów - wektora wag i 
                                       # aktualnie przetwarzanych danych treningowych - czyli liczymy sumę 1*bias+x1*w1+x2*w2

            a = activation_f(sum) # zgodnie z modelem perceptronu, obliczona suma podawana jest na funkcję aktywacji

            err =  target[i] - a  # wynik otrzymany na wyjściu funkcji aktywacji porównujemy z wartością oczekiwaną
                                  # otrzymujemy błąd - jako różnicę między nimi

            if err != 0 :   # jeśli ta różnica nie wynosi zero - czyli nie otrzymaliśmy takiej wartości jaką oczekujemy
                w = w + lr*err*x_data[i]  #aktualizujemy parametry wag 

    return w
````

````python
w = fit_perceptron(data, t)

print(w)
````

Dla pierwszego przykładu:

```
-3*1+1*1+2*1 = 0 --> activation_f(0) = 1 --> 1 (otrzymujemy poprawną wartość)
```

Dla przykładu drugiego:

```
-3*1+1*0+2*1 = -1 --> activation_f(-1) = 0 --> 0 (otrzymujemy poprawną wartość)
```

Dla pozostałych przykładów również otrzymamy poprawne wartości.

## Perceptron prosty 2

obraz_2021-04-13_135107.png

````python
import numpy as np
````

````python
x_data = np.array(
    [
     [4, 6, 1],
     [5, 4, 1],
     [6, 6, 1],
     [7, 3, 1],
     [1, 0, 1],
     [2, -1, 1],
     [3, 4, 1],
     [0, -2, 1]
    ]
)
# -1 - klasa 1, 1 - klasa 2
y_data = np.array([1,1,1,1,-1,-1,-1,-1])
````

````python
def activation_bipolar(x):
    return 1 if x >= 4 else -1
````

````python
def fit_perceptron(x_data, target): 
    w = np.zeros(len(x_data[0])) 
    lr = 1 # współczynnik uczenia
    epochs = 10 # liczba epok

    for epoch in range(epochs): # iteracja po wszystkich epokach
        for i, x in enumerate(x_data): 

            sum = np.dot(x_data[i], w)

            a = activation_bipolar(sum) 
            
            err =  target[i] - a

            if err != 0 : 
                w = w + lr*err*x_data[i]

    return w
````

````python
w = fit_perceptron(x_data, y_data)

print(w)
````

Im mniejszy learning rate, tym mniejsze wagi. Natomiast zmiana liczby epok nie wpływa na wgi, nawet zmiana tego parametru o rząd wielkości.

Dla pierwszego przykładu:

```
 2*4+4*6+(-12)*1 = 20 --> activation_b(20) = 1 --> 1 (otrzymujemy poprawną wartość)

```

Dla przykładu piątego:

```
2*1+4*0+(-12)*1 = -10 --> activation_b(-10) = -1 --> -1 (otrzymujemy poprawną wartość)
```

## Implementacja perceptronu z wykorzystaniem *Tensorflow - Keras*

Implementację perceptronu można również z powodzeniem wykonać wykorzystując bibliotekę Tensorflow oraz API Keras.

Jako zestaw danych wykorzystamy zbiór *Mnist*, który wbudowany jest w API Keras. Zawiera on obrazy 10 cyfr w rozdzielczości 28x28 pikseli. 

Wykorzystując ten zbiór i perceptron prosty, nie przeprowadzimy klasyfikacji dla 10 klas. Przygotujmy dane zatem tak, aby klasyfikator był w stanie rozróżnić czy na danym obrazie wejściowym jest cyfra 9 czy nie. 

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# przygotowanie danych 
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# zmieniamy etykiety klas: jeśli aktualnie była 9 - przyjmie wartość 1, jeśli inna - wartość 0
y_train_9 = (y_train == 9)
y_test_9 = (y_test == 9)

# zmieniamy 28x28 pikseli na 784 cechy
X_train = X_train.reshape(X_train.shape[0], 784)
X_test = X_test.reshape(X_test.shape[0], 784)

# przeskalowanie wartości z <0;255> na <0;1>
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255

#wyodrębnijmy jedną próbkę z danymi i wyświetlmy ją (później wykorzystana zostanie do sprawdzenia wyniku działania modelu)
test_example = X_test[20]
test_example_show = plt.imshow(test_example.reshape(28,28), cmap=matplotlib.cm.binary)

````

Z wykorzystaniem API Keras możemy zaimplementować model perceptronu, który wykorzystuje stochastyczne zejście gradientowe (SGD) jako metodę optymalizacji. W takim przypadku nie możemy wykorzystać podstawowej funkcji aktywacji w perceptronie (progowej unipolarnej) ze względu na to, iż nie jest ona ciągła (i różniczkowalna) dla wartości 0. W przypadku optymalizacji SGD - wagi aktualizowane są dla niezerowych gradientów, czego nie można osiągnąć wykorzystując jako funkcję aktywacji funkcję progową niesymetryczną. 

Bardzo zbliżoną funkcją do funkcji krokowej jest tak zwana funkcja "hard sigmoid", której implementacja dostępna jest w API Keras (ma właściwości funkcji sigmoidalnej, lecz zmienia się wolniej; jest różniczkowalna). 

image.png

````python
from tensorflow.keras.activations import hard_sigmoid

# tworzymy perceptron z wykorzystaniem API Keras

# Sequential - grupowanie liniowe warstw
model = Sequential()
# w perceptronie jest tylko jedna 'warstwa'; dodajemy warstwę gęstą: 1 - wymiarowość wyniku, 
# input_shape - liczba wejść
# activation - funckcja aktywacji; kernel_initializer - inicjalizacja niezerowych wag początkowych
model.add(Dense(1, input_shape=(784,), activation=hard_sigmoid, kernel_initializer='glorot_uniform'))

# loss - funkcja straty (standardowa dla problemów klasyfikacji binarnej)
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])

#wyświetlenie wag przed uczeniem
print(model.weights)

# trenowanie modelu na przygotowanych danych treningowych (100 epok, wielkość patrii = 25)
model.fit(X_train, y_train_9, epochs=100, batch_size=32)

# ewaluacja modelu - dokładność na danych testowych
_, accuracy = model.evaluate(X_test, y_test_9)
print(accuracy)

#wyświetlenie wag po uczeniu
print(model.weights)


````
